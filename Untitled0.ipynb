```python
import torch
from torch import nn, optim, tensor
import matplotlib.pyplot as plt
import torchdiffeq as tdf
import numpy as np
from IPython.display import clear_output
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
class growth_func(nn.Module):
    def __init__(self, *args):
        super(growth_func, self).__init__()
        self.K = args[0]
        self.r = args[1]

    def forward(self,t,y):
        return tensor(self.r * (1-y/self.K) * y)


t = torch.linspace(0.0,20,1000)
true_y = tdf.odeint(growth_func(100,0.95), tensor(0.01),t)
plt.plot(true_y)
```

```python
def get_batch(batch_time, batch_num, total_time=len(t)):
    s = torch.from_numpy(np.random.choice(np.arange(total_time - batch_time), batch_num, replace=False))
    batch_y0 = torch.Tensor([[i] for i in true_y[s]])
    batch_t = t[:batch_time]
    batch_y = torch.stack([true_y[s+i] for i in range(batch_time)],dim=0)
    return batch_y0.to(device), batch_t.to(device), torch.Tensor(batch_y).to(device)
```

```python
class ODEFunc(nn.Module):
    def __init__(self):
        super(ODEFunc, self).__init__()

        self.net = nn.Sequential(
            nn.Linear(1,50),
            nn.Tanh(),
            nn.Linear(50,50),
            nn.Tanh(),
            nn.Linear(50,1)
        )

        for m in self.net.modules():
            if isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, mean=0, std=0.1)
                nn.init.constant_(m.bias, val=0)

    def forward(self, t, y):
        return self.net(y)
```

```python
def plot_traj(pred_y, true_y):
    f, ax = plt.subplots(1,1, figsize=(6,5))
    ax.plot(t.detach().numpy(), true_y.detach().numpy() )
    ax.plot(t.detach().numpy(), pred_y.detach().numpy().squeeze(), linestyle="--")
    plt.show()
```

```python
func = ODEFunc().to(device)
optimizer = optim.Adam(func.parameters(), lr=1e-3)
loss_max = 100
n = 100
```

```python
for i in range(100000):
    optimizer.zero_grad()
    batch_y0, batch_t, batch_y = get_batch(20,100)
    pred_y = tdf.odeint(func, batch_y0, batch_t)


    loss = torch.mean(torch.abs(pred_y.squeeze() - batch_y))
    loss.backward()
    if loss < 1e-2:
      break
    optimizer.step()
    # print(loss)
    # with torch.no_grad():
        # y_pred = tdf.odeint(func, torch.Tensor([[0.01]]), t)

    # if loss < loss_max:
    #   loss_max = loss.item()
    #   print(loss_max)

    if i % 100 == 0:
      print("loss: {} at {}".format(loss.item(),i))
    # plot_traj(y_pred, true_y)
    # clear_output(wait=True)


```
